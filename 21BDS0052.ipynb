2 Way Contingency table

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['signs_of_mental_illness'], df['body_camera'])
print("Contingency Table (Signs of Mental Illness vs Body Camera):\n", contingency_table)

# signs of mental illness
row_profile = contingency_table.div(contingency_table.sum(axis=1), axis=0)
print("\nRow Profile Dataset:\n", row_profile)

# body camera
column_profile = contingency_table.div(contingency_table.sum(axis=0), axis=1)
print("\nColumn Profile Dataset:\n", column_profile)

# Relative frequency
relative_frequency = contingency_table.div(contingency_table.sum().sum())
print("\nRelative Frequency Dataset:\n", relative_frequency)

#Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("\nChi-square Test Results:")
print(f"Chi2 Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)
```

```python
# Three-way contingency table
three_way_table = pd.crosstab(index=[df['signs_of_mental_illness'], df['body_camera']], columns=df['threat_level'])
print("Three-Way Contingency Table:\n", three_way_table)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['threat_level'] = label_encoder.fit_transform(df['threat_level'])

K_means = df[['age', 'threat_level']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(K_means)

inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-', markersize=8)
plt.xlabel('Number of Clusters, K')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.title('Elbow Method for Optimal K')
plt.grid()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df['signs_of_mental_illness'] = df['signs_of_mental_illness'].astype(str).map({'True': 1, 'False': 0})

df_cluster = df[['age', 'signs_of_mental_illness']].dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster)

optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
clusters = kmeans.fit_predict(X_scaled)
df_cluster['Cluster'] = clusters

plt.figure(figsize=(10, 6))
plt.scatter(df_cluster['age'], df_cluster['signs_of_mental_illness'],
            c=df_cluster['Cluster'], cmap='viridis', marker='o',
                      s=50, alpha=0.6, edgecolor='k')

centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], color='red', marker='X', s=200, label='Centroids')

plt.xlabel('Age')
plt.ylabel('Signs of Mental Illness (Encoded)')
plt.title('K-means Clustering of Age and Signs of Mental Illness')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import numpy as np

numeric_data = df.select_dtypes(include=[np.number])

imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(numeric_data)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(imputed_data)

pca = PCA()
pca_result = pca.fit(scaled_data)

eigenvalues = pca.explained_variance_ratio_
loadings = pca.components_

print("Eigenvalues:", eigenvalues)
print("Loadings:\n", loadings)

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, marker='o', linestyle='--')
plt.title("Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Variance Explained")
plt.show()
```

```python
n_comp = 2
pca_selected = PCA(n_components=n_comp)
pca_selected_result = pca_selected.fit_transform(scaled_data)

plt.figure(figsize=(10, 7))
plt.scatter(pca_selected_result[:, 0], pca_selected_result[:, 1], alpha=0.5)


for i, feature in enumerate(numeric_data.columns[:n_comp]):
    plt.arrow(0, 0, pca_selected.components_[0, i], pca_selected.components_[1, i],
                  color='r', alpha=0.7)
plt.text(pca_selected.components_[0, i] * 1.2, pca_selected.components_[1, i] * 1.2,
feature, color='g', ha='center', va='center')

plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("Biplot (First 2 Principal Components)")
plt.grid()
plt.show()

print("\nExplained Variance Ratio:", pca_selected.explained_variance_ratio_)
print("\nPrincipal Components:\n", pca_selected.components_)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import LabelEncoder

file_path = "shootings.csv"
data = pd.read_csv(file_path)

features = ['manner_of_death', 'armed', 'gender', 'race', 'signs_of_mental_illness',
            'threat_level', 'flee', 'body_camera', 'arms_category']
target = 'age'

data_clean = data.dropna(subset=[target])

for col in features:
    if data_clean[col].dtype == 'object':
        data_clean[col].fillna('Unknown', inplace=True)

label_encoders = {}
for col in features:
    if data_clean[col].dtype == 'object':
        le = LabelEncoder()
        data_clean[col] = le.fit_transform(data_clean[col])
        label_encoders[col] = le

X = data_clean[features]
y = data_clean[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R² Score: {r2}")
```
