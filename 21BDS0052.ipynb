```python
mean_age_by_race = df_deduped.groupby('race')['age'].mean()
print(mean_age_by_race)
```

```python
count_by_gender = df_deduped['gender'].value_counts()
print(count_by_gender)
```

```python
contingency_table = pd.crosstab(df_deduped['race'], df_deduped['gender'])
print(contingency_table)
sns.heatmap(contingency_table, annot=True, cmap='coolwarm', linewidths=1)
plt.title('Contingency Table: Race vs Gender')
plt.show()
```

```python
sns.FacetGrid(df).map(sns.histplot, "age").add_legend()
plt.show()
```

```python
sns.distplot(df['age'], kde=False)
plt.show()
```

```python
sns.boxplot(x="age", data=df)
plt.show()
```

```python
sns.stripplot(x='race', y='age', data=df_deduped, jitter=True)
plt.title('Dot Chart: Age by Race')
plt.show()
```

```python
plt.scatter(df["age"], df["signs_of_mental_illness"].apply(lambda x: 1 if x else 0))
plt.xlabel("Age")
plt.ylabel("Signs of Mental Illness (1 = True, 0 = False)")
plt.show()
```

```python
df['signs_of_mental_illness_numeric'] = df['signs_of_mental_illness'].apply(lambda x: 1 if x else 0)
df['threat_level_numeric'] = df['threat_level'].astype('category').cat.codes
sns.pairplot(df, vars=['age', 'signs_of_mental_illness_numeric', 'threat_level_numeric'], kind="reg")
plt.show()
```

```python
corr = stats.pearsonr(df["age"], df["signs_of_mental_illness"].apply(lambda x: 1 if x else 0))
print("p-value:\t", corr[1])
print("cor:\t\t", corr[0])
```

```python
correlation = df[['age', 'signs_of_mental_illness']].apply(lambda x: x.astype(float)).corr(method='pearson')
sns.heatmap(correlation, annot=True, xticklabels=correlation.columns, yticklabels=correlation.columns)
plt.show()
```

```python
correlation_matrix = df[['age', 'signs_of_mental_illness_numeric', 'threat_level_numeric']].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Convert 'date' to datetime and set as index
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y')
df.set_index('date', inplace=True)
```

```python
df = df.sort_index()
df.head()
```

```python
df['age'].plot()
plt.title('Age Over Time')
plt.ylabel('Age')
plt.show()
```

```python
monthly_age = df['age'].resample('M').mean()
monthly_age.plot()
plt.title('Monthly Average Age')
plt.ylabel('Age')
plt.show()
```
