```python
plt.scatter(df["age"], df["signs_of_mental_illness"].apply(lambda x: 1 if x else 0))
plt.xlabel("Age")
plt.ylabel("Signs of Mental Illness (1 = True, 0 = False)")
plt.show()
```

```python
df['signs_of_mental_illness_numeric'] = df['signs_of_mental_illness'].apply(lambda x: 1 if x else 0)
df['threat_level_numeric'] = df['threat_level'].astype('category').cat.codes
sns.pairplot(df, vars=['age', 'signs_of_mental_illness_numeric', 'threat_level_numeric'], kind="reg")
plt.show()
```

```python
corr = stats.pearsonr(df["age"], df["signs_of_mental_illness"].apply(lambda x: 1 if x else 0))
print("p-value:\t", corr[1])
print("cor:\t\t", corr[0])
```

```python
correlation = df[['age', 'signs_of_mental_illness']].apply(lambda x: x.astype(float)).corr(method='pearson')
sns.heatmap(correlation, annot=True, xticklabels=correlation.columns, yticklabels=correlation.columns)
plt.show()
```
